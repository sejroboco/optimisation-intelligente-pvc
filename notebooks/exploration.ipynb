```python
import pandas as pd
```

```python
communes = pd.read_csv("../data/villes_france.csv")
communes.head()
```

```python
communes.shape
```

```python
communes.info()
```

```python
import folium

# Centrer la carte sur la France (coordonnées approximatives)
center_lat = communes["latitude"].mean()
center_lon = communes["longitude"].mean()
m = folium.Map(location=[center_lat, center_lon], zoom_start=6)

# Ajouter les villes
for _, row in communes.iterrows():
    folium.Marker(
        location=[row["latitude"], row["longitude"]],
        popup=row["ville"],
        icon=folium.Icon(color="blue", icon="info-sign")
    ).add_to(m)

# Tracer des lignes entre chaque paire de villes
for i in range(len(communes)):
    for j in range(i+1, len(communes)):  # j=i+1 pour éviter doublons
        coord1 = [communes.iloc[i]["latitude"], communes.iloc[i]["longitude"]]
        coord2 = [communes.iloc[j]["latitude"], communes.iloc[j]["longitude"]]
        folium.PolyLine(locations=[coord1, coord2], color="gray", weight=1, opacity=0.5).add_to(m)

m
```
